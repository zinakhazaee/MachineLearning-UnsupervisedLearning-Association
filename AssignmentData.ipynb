```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_excel('/content/drive/MyDrive/datasets_sample/market-basket-analysis/AssignmentData.xlsx')
```

```python
df.head()
```

```python
df.shape
```

```python
df.groupby('Country').size()
```

```python
basket_Germany = (df[df['Country']=='Germany'].groupby(['BillNo','Itemname'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('BillNo')
          )
basket_United_Kingdom = (df[df['Country']=='United Kingdom'].groupby(['BillNo','Itemname'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('BillNo')
          )
basket_France = (df[df['Country']=='France'].groupby(['BillNo','Itemname'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('BillNo')
          )
basket_Netherlands = (df[df['Country']=='Netherlands'].groupby(['BillNo','Itemname'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('BillNo')
          )
basket_Spain = (df[df['Country']=='Spain'].groupby(['BillNo','Itemname'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('BillNo')
          )
```

```python
def encode_units(x):
  if x <= 0:
    return 0
  if x >= 1 :
    return 1
```

## Germany ,United Kingdom ,France ,Netherlands ,Spain

#Frequent Items

```python
basket_set_Germany = basket_Germany.applymap(encode_units)
frequent_items_Germany = apriori(basket_set_Germany , min_support= 0.07 , use_colnames= True)
frequent_items_Germany.head()
```

```python
basket_set_United_Kingdom = basket_United_Kingdom.applymap(encode_units)
frequent_items_United_Kingdom = apriori(basket_set_United_Kingdom , min_support= 0.06 , use_colnames= True)
frequent_items_United_Kingdom.head(10)
```

```python
basket_set_France = basket_France.applymap(encode_units)
frequent_items_France = apriori(basket_set_France , min_support= 0.07 , use_colnames= True)
frequent_items_France.head()
```

```python
basket_set_Netherlands = basket_Netherlands.applymap(encode_units)
frequent_items_Netherlands = apriori(basket_set_Netherlands , min_support= 0.07 , use_colnames= True)
frequent_items_Netherlands.head()
```

```python
basket_set_Spain = basket_Spain.applymap(encode_units)
frequent_items_Spain = apriori(basket_set_Spain , min_support= 0.07 , use_colnames= True)
frequent_items_Spain.head()
```

#Rules

```python
rules_Germany = association_rules(frequent_items_Germany, metric='lift',min_threshold= 1)
best_rules_Germany = rules_Germany[(rules_Germany['lift'] >= 0.6) & (rules_Germany['confidence'] >= 0.6)]
best_rules_Germany.sort_values(by='confidence', ascending=False)
```

```python
rules_France = association_rules(frequent_items_France, metric='lift',min_threshold= 1)
best_rules_France = rules_France[(rules_France['lift'] >= 0.6) & (rules_France['confidence'] >= 0.6)]
best_rules_France.sort_values(by='confidence', ascending=False)
```

```python
rules_Netherlands = association_rules(frequent_items_Netherlands, metric='lift',min_threshold= 1)
best_rules_Netherlands = rules_Netherlands[(rules_Netherlands['lift'] >= 0.6) & (rules_Netherlands['confidence'] >= 0.6)]
best_rules_Netherlands.sort_values(by='confidence', ascending=False)
```

```python
rules_Spain = association_rules(frequent_items_Spain, metric='lift',min_threshold= 1)
best_rules_Spain = rules_Spain[(rules_Spain['lift'] >= 0.6) & (rules_Spain['confidence'] >= 0.6)]
best_rules_Spain.sort_values(by='confidence', ascending=False)
```
