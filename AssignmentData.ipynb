```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_excel('/content/drive/MyDrive/datasets_sample/market-basket-analysis/AssignmentData.xlsx')
```

```python
df.head()
```

```python
df.shape
```

```python
basket = (df.groupby(['BillNo','Itemname'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('BillNo')
          )
basket
```

```python
def encode_units(x):
  if x <= 0:
    return 0
  if x >= 1 :
    return 1
```

```python
basket_set = basket.applymap(encode_units)
frequent_items = apriori(basket_set , min_support= 0.07 , use_colnames= True)
frequent_items
```

```python
rules = association_rules(frequent_items , metric='lift',min_threshold= 1)
rules
```

```python
best_rules = rules[(rules['lift'] >= 0.6) &
      (rules['confidence'] >= 0.6)]
```

```python
best_rules.sort_values(by='confidence', ascending=False)
```
